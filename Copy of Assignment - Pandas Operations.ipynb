# <b>Problem 5 : Print the company name which manufactured the most expensive car. Also print the price of the most expensive car

```python
# Your code here
import pandas as pd

automobile_df = automobile_df.replace({'?': np.NaN, 'n.a': np.NaN})
automobile_df['price'] = pd.to_numeric(automobile_df['price'])
highest_price = max(automobile_df['price'])
make_of_most_expensive_car = automobile_df[automobile_df['price'] == highest_price]['make'].to_string(index=False).strip()

print(f"The most expensive car is made by {make_of_most_expensive_car} and costs ${highest_price:.2f}")
```

# <b> Problem 8 : Create a new column for the updated price. This updated price will be calculated on top of the given price - if the engine is in front, price will be same else if the engine is in rear, price will be doubled </b>


```python
# Your code here. Use apply and lambda
import numpy as np
import pandas as pd

def revised_price(engine_location,price):
    if engine_location == "front":
        updated_price = price
    else:
        updated_price = 2 * price
    return updated_price

# Assuming you have an existing DataFrame called "automobile_df"
# and a column named "price"
automobile_df["updated_price"] = automobile_df.apply(lambda x: revised_price(x["engine-location"],x["price"]), axis=1)
```

# <b>Problem 11 : Concatenate the two data frames given below firstly row wise and secondly column wise.</b>
