## <u>Problem 1</u>

A list contains the average daily temperature(in degree Celsius) of a city over a particular week. Write a Python code to swap the highest and the lowest temperatures

```python
# A list containing average daily temperature over a week
temperatures = [34, 40, 29, 33, 42, 37, 39 ]

# The expected output
# output_temperatures = [34, 40, 42, 33, 29, 37, 39]
```

```python
# Store the highest temperature
temperatures = [34, 40, 29, 33, 42, 37, 39 ] #Enter input list
max_temp = max(temperatures)  #Using max function for list
print("Highest temperature is : " +str(max_temp)) #printing max temperature

# Index of the element with the highest temperature
max_temp_index =temperatures.index(42) #To get index element used index method
print("Index of the element with the highest temperature is : " +str(max_temp_index))
```

```python
# Store the lowest temperature
temperatures = [34, 40, 29, 33, 42, 37, 39 ] #Enter input list
min_temp = min(temperatures)  #Using min function for list
print(min_temp) #printing max temperature

# Index of the element with the lowest temperature
min_temp_index =temperatures.index(29)  #To get index element used index method
print(min_temp_index)
```

```python
# Swap the highest and the lowest temperatures
temperatures = [34, 40, 29, 33, 42, 37, 39 ]
temperatures[max_temp_index],  temperatures[min_temp_index] = temperatures[min_temp_index],temperatures[max_temp_index] #swapping values for max and min temperature using assign operator
print(temperatures)
```

```python
# Print the output list
(output_temperatures) = (temperatures) #assign output value to new variable
print(output_temperatures)
```

### Measures of Averages

* Mean of a group is defined as the sum of the elements in the group divided by the number of elements in the group.
* Median is the middle number in a <b>sorted</b>, ascending or descending, list of numbers. If the list has $N$ numbers where $N$ is odd, then median is the element in the middle i.e, $(\frac{N+1}{2})^{th}$ element. If $N$ is even, then median is the mean of $(\frac{N}{2})^{th}$ and $(\frac{N}{2}+1)^{th}$ elements

```python
# List of product prices

prod_price_list = [400, 250, 800, 550, 600, 820, 720, 15000, 360,250]
```

```python
# Calculate the mean of the product prices. Use list functions. Also print the mean_price
prod_price_list = [400, 250, 800, 550, 600, 820, 720, 15000, 360,250] #Enter input list
sum_of_price= sum(prod_price_list) #Used sum function to get sum of elements in list
num_of_elements = len(prod_price_list) #used length function to get length of elements in list
mean_price = (sum_of_price/num_of_elements)
print("mean price is : " +str(mean_price))
```

```python
# Use the corresponding formula to calculate the median and print the median
prod_price_list = [400, 250, 800, 550, 600, 820, 720, 15000, 360,250]
mid = len(prod_price_list) // 2 #calculated mid of list
median = (prod_price_list[mid] + prod_price_list[~mid]) / 2 #calculated median using mid value and ~mid value
print("Median of list is : " + str(median))
```

```python
# Check which is greater, mean or median
print(mean_price)
print(median)
if mean_price > median : #Used if else condition to get greater value using comparision operator
    print ("mean is greater than median")
else :
    print ("median is greater than mean")
```

## <u> Problem 3 </u>

Having a nested list sometimes might be a bit problematic. An individual was asked to collect the names of companies in the technology sector. While creating the list, by mistake the last three companies were subsumed in a list as shown below. You are required to get rid of the nesting

```python
# The list of tech companies curated by the individual
tech_companies = ['Qualcomm','Google','Apple',['Nvidia','Cisco','Samsung']]

# The final list that we want
# correct_tech_companies = ['Qualcomm','Google','Apple','Nvidia','Cisco','Samsung']
```

```python
# Write your code below
correct_tech_companies1 = (tech_companies[0:3])#Used list indexing to get first 3 values
correct_tech_companies2 = (tech_companies[3][0:]) #Used list indexing to get first 3 values
correct_tech_companies =correct_tech_companies1 + correct_tech_companies2[0:] #concatenate 2 lists
print(correct_tech_companies)
```

```python
# Do the same task another way
correct_tech_companies1 = (tech_companies[-4:-1]) #Used list indexing to get first 3 values
correct_tech_companies2 = (tech_companies[-1][-3:]) #Used list indexing to get first 3 values
correct_tech_companies =correct_tech_companies1 + correct_tech_companies2[-3:] #concatenate 2 lists
print(correct_tech_companies)
```
