## <u>Problem 1</u>

### Another measure of average

* Mode is defined as the value that appears most often in a set of data values. Mode corresponds to the most frequent value.
* You are given a list of numbers and you need to calculate the mode of this list

```python
# A list containing average daily temperature over 10 days
temperatures = [34, 40, 29, 33, 42, 40, 39, 34, 34,33 ]
```

```python
# Step 1 : Find the unique set of values. Store this in a list
unique_temperatures1 = set(temperatures)
unique_temperatures=list(unique_temperatures1)
```

```python
# Print the unique temperature list
print(unique_temperatures)
```

```python
# Step 2 : Create an empty dictionary which will contain the unique temperatures as keys and their counts in the original temperatures list as values
temp_dict = {}
temp_dict1 = list(set(temperatures))


# Set the values for each key separately
# First key
first_key=temperatures.count(33)

# Second key
second_key=temperatures.count(34)

# Third key
third_key=temperatures.count(39)

# Fourth key
fourth_key=temperatures.count(40)

# Fifth key
fifth_key=temperatures.count(42)

# Sixth key
sixth_key=temperatures.count(29)

count_values=[first_key,second_key,third_key,fourth_key,fifth_key,sixth_key]
temperatures_and_counts = list(zip(temp_dict1,count_values))
temperatures_and_counts_dict=dict(temperatures_and_counts)
print(temperatures_and_counts_dict)
```

```python
# Step 3: Store the dictionary values in a list
temp_dict_frequencies=[first_key,second_key,third_key,fourth_key,fifth_key,sixth_key]

# Print the list of dictionary values
print(temp_dict_frequencies)

# Find the index of the most frequent temperature value
from collections import Counter
temperature_counts = Counter(temperatures)
most_frequent_temperature = max(temperature_counts, key=temperature_counts.get)
index_of_most_frequent_temperature = temperatures.index(most_frequent_temperature)

# Print the index
print(index_of_most_frequent_temperature)
```

```python
# Step 4 : Create the list of dictionary keys
temp_dict_keys = temp_dict1

# Print the list of dictionary keys
print(temp_dict_keys)

# Calculate the mode temperature
mode_temperature = temperature_counts.most_common(1)[0][0]
print(mode_temperature)
```
